```
import numpy as np
import pandas as pd
import requests
import bs4
import lxml.etree as xml
```

```
my_session = requests.session()
for_cookies = my_session.get("https://www.hespress.com")
cookies = for_cookies.cookies
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; Win64; x64; rv:57.0) Gecko/20100101 Firefox/57.0'}
my_url = 'https://www.hespress.com'

response = my_session.get(my_url, headers=headers, cookies=cookies)
res = response.text
web_page = bs4.BeautifulSoup(res, "lxml")
```

```
latest_news = web_page.find_all(name="h3" ,attrs={"class":"latest_news_title"})
href_latest_news = []
latest_news_text = []

for elem in latest_news:
  link = elem.find(name="a")
  href_latest_news.append(link.attrs["href"])               
  link = link.text.strip('\n ')
  latest_news_text.append(link)
  
latest_news_text
```

```
#get Articles 
base_link = 'https://www.hespress.com'
all_Articles = []
for elem in href_latest_news:
  lien = base_link+elem
  response = my_session.get(base_link+elem, headers=headers, cookies=cookies)
  res = response.text
  web_page = bs4.BeautifulSoup(res, "lxml")
  getArticle = web_page.find(name="div" , attrs={"id":"article_body"}).find_all(name="p")
  paragraph = ""
  for elem in getArticle:
    paragraph = paragraph + elem.text
  
  all_Articles.append(  (lien,paragraph)  )

  paragraph = ""

all_Articles
```
